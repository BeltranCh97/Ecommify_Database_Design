After successfully mounting your drive, you can now load your CSV files into a pandas DataFrame. Remember to replace `'path/to/your/file.csv'` with the actual path to your CSV file in Google Drive (e.g., `/content/drive/MyDrive/your_folder/your_file.csv`).

```python
import os
import pandas as pd

# Ruta a la carpeta en Google Drive
folder_path = '/content/drive/MyDrive/Maestria en ingenieria de software/TallerBaseDatosUnidad1/'

if not os.path.exists('/content/drive/MyDrive'):
    print("ERROR: Google Drive no parece estar montado. Por favor, ejecuta la celda de drive.mount y autoriza el acceso.")
elif not os.path.exists(folder_path):
    print(f"ERROR: La carpeta no existe en la ruta: {folder_path}")
else:
    try:
        # Listar archivos CSV
        files = [f for f in os.listdir(folder_path) if f.lower().endswith('.csv')]

        if not files:
            print("No se encontraron archivos CSV en el directorio especificado.")
        else:
            for file_name in files:
                file_path = os.path.join(folder_path, file_name)
                print(f"\n{'='*60}")
                print(f"ANALIZANDO: {file_name}")
                print(f"{'='*60}")

                # Cargar y analizar
                df_temp = pd.read_csv(file_path)

                print("\n--- INFORMACIÓN (df.info) ---")
                df_temp.info()

                print("\n--- ESTADÍSTICAS (df.describe) ---")
                display(df_temp.describe())

    except Exception as e:
        print(f"Ocurrió un error inesperado: {e}")
```
